### Задача 10.2. Правила Английской Капитализации

В англоязычном медиапространстве в некоторых специфических случаях принято писать текст в формате [***Title Case***](https://ru.wikipedia.org/wiki/%D0%97%D0%B0%D0%B3%D0%BE%D0%BB%D0%BE%D0%B2%D0%BE%D1%87%D0%BD%D1%8B%D0%B9_%D1%80%D0%B5%D0%B3%D0%B8%D1%81%D1%82%D1%80) — это когда каждое слово записывается с заглавной буквы (а не только первое слово в предложении и не только имена собственные, как в русском). Например, *title case*’ом обычно записываются названия произведений искусства: альбомов (как «A Moon Shaped Pool» Radiohead), фильмов (как «Dr. Strangelove or: How I Learned to Stop Worrying and Love the Bomb» Стэнли Кубрика), картин (как [«The Storm on the Sea of Galilee»](https://www.singulart.com/blog/wp-content/uploads/2023/08/image-68-1140x1417.png) Рембрандта) и так далее. Кроме того, особенно в США так часто записываются заголовки в СМИ (например, [«Stop Trying to Make the Humanities Relevant»](https://www.theatlantic.com/ideas/2025/12/humanities-crisis-ai-camus/685233/) в газете The Atlantic).
