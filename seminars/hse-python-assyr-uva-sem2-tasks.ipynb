### Задача 2.2. Волшебное слово

В питоне есть много волшебных слов — `print()`, `input()` и других, а в жизни нам часто нужно другое волшебное слово — «пожалуйста». В этой задаче вы напишете программу, которая добавляет его в разные просьбы, чтобы они звучали вежливее.

Пользователь вводит строку с просьбой (например, *приходи поскорее*).
- Если эта просьба заканчивается вопросительным знаком (`"?"`), то нужно напечатать: `"Это не просьба, а вопрос."` — и больше ничего не делать.
- Если это всё же просьба, то нужно проверить, не содержится ли уже в этой просьбе слово «пожалуйста».

    - Если слово «пожалуйста» уже есть в просьбе, то нужно просто напечатать просьбу в неизменном виде.
    - Если в итоге вы действительно напечатали просьбу, в которой есть слово «пожалуйста» (вне зависимости от того, добавили ли его вы или оно было там изначально), то надо также написать на другой строке: `"Готово."`.
