### Задача 6.4. Аккадский парсер

Как и в русском языке, в аккадском есть прилагательные, которые согласуются по роду, числу и падежу со своим существительным с помощью суффиксов. В этой задаче вам предстоит написать программу, которая будет проводить морфологический разбор аккадских прилагательных. Знание аккадского для решения задачи не требуется!

- Дан словарь `vocab`, ключами которого являются корни прилагательных аккадского языка, а значениями — русские переводы
- Ваша программа получает на вход слово, которое может быть (а может и не быть) настоящим аккадским прилагательным в каком-то падеже, числе и роде
- Задача программы — определить, является ли это слово аккадским прилагательным
    - Если да — указать соответствующее значение (из словаря) и грамматические признаки (род и падеж)
    - Если нет — вывести сообщение о том, что такого слова быть не может

Немного про аккадскую морфологию (очень упрощённо в рамках задания):

- Прилагательные в аккадском устроены так: **корень** + **суффикс женского рода** (если нужен) + **суффикс падежа**. Если прилагательное стоит в мужском роде, после корня сразу идёт суффикс падежа (у мужского рода «нулевой» суффикс). Форм без суффикса падежа не бывает
- Форма суффикса женского рода зависит от корня прилагательного. Если корень оканчивается на сочетание из двух согласных, то суффикс женского рода звучит как -*at*, а если на что угодно другое — то как -*t*. Например: прилагательное *yap* ‘красивый’ в женском роде будет звучать как *yap*-***t***-*um*, а прилагательное *dašp* ‘медовый’ — как *dašp*-***at***-*um*. (Без дополнительного гласного *a* произношение сочетания из трёх согласных было бы слишком неудобным!) Если выбрана неправильная форма суффикса женского рода, то мы считаем, что прилагательное образовано неправильно
- В этом задании будем рассматривать только формы в трёх падежах: именительном, родительном и винительном (их окончания приведены ниже), и только в единственном числе

Вот все возможные окончания:

|         | жен.       | муж.     |
|---------|------------|----------|
| им. п.  | -*(a)t-um* | -*um*    |
| род. п. | -*(a)t-im* | -*im*    |
| вин. п. | -*(a)t-am* | -*am*    |

Вот примеры правильной работы программы. В четвёртом и пятом случае неправильно образована форма женского рода (выбрана неверная форма суффикса), в шестом случае программа не может найти в словаре несуществующий корень `"yapampimpumpam"`.

```

> yaptim
```
