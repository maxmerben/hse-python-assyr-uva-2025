### Задача 5.1. Оглавление

В этой задаче вам предстоит заняться издательским делом — красиво оформить оглавление для выпуска журнала, который публикует научные статьи.

Даны три списка — список с названиями статей в определённом выпуске, список с названиями автор:ок этих статей и список, в котором для каждой статьи указан номер первой страницы. Списки отсортированы в порядке появления статей в выпуске. Гарантируется, что списки составлены корректно — а значит, количество элементов во всех трёх списках одинаково.

Программа должна напечатать красивое оглавление в следующем формате:
- Каждая строка содержит данные по определённой статье
- Строка начинается с имени автор:ки, затем ставится точка и пробел, затем — название статьи
    - С названием статьи есть небольшая сложность. Названия статей часто состоят из двух частей, разделённых **двоеточием и пробелом** — в таких случаях обычно первая часть представляет собой некоторое завлекательное выражение (зачастую с цитатой или юмором), а вторая — непосредственно «серьёзное» название статьи ([например](https://www.ceeol.com/search/article-detail?id=587224), «*Косточка сладкая, на чужбине я…*: поэзия таджикских трудовых мигрантов в России» [или](https://cyberleninka.ru/article/n/reduplitsirovatsya-to-reduplitsirovalsya-da-ne-vyreduplitsirovalsya-konstruktsii-s-povtorom-glagola-v-ashtynskom-darginskom) «Редуплицироваться-то редуплицировался, да не выредуплицировался: конструкции с повтором глагола в аштынском даргинском»). В таких случаях в оглавлении необходимо вывести ***только* «серьёзное» название**, то есть только вторую часть.
- После названия печатаются нижние подчёркивания, а затем — номер первой страницы статьи. Количество нижних подчёркиваний должно быть таким, чтобы общая длина строки (вместе с именем автор:ки, названием, подчёркиваниями и номером страницы) равнялась 120 символам. (Это облегчит читатель:ницам поиск нужной страницы.)
