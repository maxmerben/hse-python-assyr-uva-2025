*Комментарий*. Конечно, это учебная задача, и в подсчёте, который в ней нужно провести, не особо много научного смысла — к примеру, разумнее было бы сравнивать не количество букв, а количество *звуков* в словах. Однако в ней есть некоторая связь с реальными исследованиями. Если вы решили задачу правильно, то могли заметить по получившимся данным, что количество единиц в словах как будто бы отрицательно коррелирует с количеством слов в предложении — то есть чем больше сами слова, тем меньше слов нужно, чтобы выразить определённый смысл. Это отражает различия в морфологической структуре языков — [есть языки](https://ru.wikipedia.org/wiki/%D0%9F%D0%BE%D0%BB%D0%B8%D1%81%D0%B8%D0%BD%D1%82%D0%B5%D1%82%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B9_%D1%8F%D0%B7%D1%8B%D0%BA), где слова бывают огромные, и такие слова могут выражать то, для чего в других языках требуется целое предложение, а [есть языки](https://ru.wikipedia.org/wiki/%D0%98%D0%B7%D0%BE%D0%BB%D0%B8%D1%80%D1%83%D1%8E%D1%89%D0%B8%D0%B9_%D1%8F%D0%B7%D1%8B%D0%BA), где слова маленькие, и чтобы выразить тот же смысл, их нужно много. Например, известный лингвист XX века Джозеф Гринберг так же [пытался](https://publications.hse.ru/pubs/share/folder/rdm1tqlpee/70660577.pdf#page=8) показать эту разницу вычислениями, только он считал не количество букв или звуков, как вы в этой задаче, а количество морфем на слово. (Но, разумеется, количество морфем и количество букв коррелируют между собой, поэтому ваши результаты тоже выглядят в некоторой степени научно осмысленными.)
