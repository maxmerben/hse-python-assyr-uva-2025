### Задача. Регулярные выражения

Напишите функцию **`find_comments()`** для поиска в литературном тексте нумерованных комментариев-сносок.

В оцифрованном литературном тексте книги есть комментарии-сноски. Все такие комментарии заключены в квадратные скобки, а некоторые из них также пронумерованы. Нумерованные комментарии — это такие, перед которыми (внутри скобок) стоит номер и пробел. Например, в этом отрывке есть один нумерованный комментарий и один ненумерованный:

> - Avant tout dites-moi, comment vous allez, ch&#232;e amie?*[25 Прежде всего скажите, как ваше здоровье, милый друг?]* Успокойте меня, - сказал он, не изменяя голоса и тоном, в котором из-за приличия и участия просвечивало равнодушие и даже насмешка. *[Рукопись обрывается.]*

Задача функции — найти все нумерованные комментарии. К примеру, если подать в неё текст выше, она вернёт строку «*25 Прежде всего скажите, как ваше здоровье, милый друг?*», а второй, ненумерованный комментарий не вернёт.

Функция должна:

- Принимать на вход текст (типа `str`), в котором нужно найти все нумерованные комментарии
- Возвращать список, состоящий из всех нумерованных комментариев (типа `str`)
