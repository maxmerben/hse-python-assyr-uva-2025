## Вопрос 9

Выберите из вариантов ниже регулярное выражение, которое как можно более точно найдёт в случайном русском тексте все слова с корнем «био» (например, *биоробот*, *пробиотики* и так далее):

1. `"(а-яё+)био(а-яё+)"`
2. `"[а-яё]*био[а-яё]*"`
3. `"[а-яё]*|био|[а-яё]*"`
4. `"[а-яё]?[био][а-яё]?"`

___

(Правильный ответ: 2.)

```python
### ТЕСТИРОВАНИЕ ###

import re

text = "Это лобио, биоробот, пробиотики — ибо нефиг."
for regex in (
    "(а-яё+)био(а-яё+)",
    "[а-яё]*био[а-яё]*",    # <-- то что нужно
    "[а-яё]*|био|[а-яё]*",
    "[а-яё]?[био][а-яё]?"
):
    print(re.findall(regex, text))
```

```python
### РЕШЕНИЕ ###

import re

def find_questions(text):
    spisok = re.findall("[^\.\?!]+\?+!*", text)
    for i in range(len(spisok)):
        spisok[i] = spisok[i].strip()
    return spisok
```

- Удалите из списка токенов стоп-слова (используйте список стоп-слов из `nltk`)
- Удалите из списка токенов знаки препинания (подсказка: это можно сделать с помощью регулярных выражений)
- Постройте частотный словарь, в котором каждому слову соответствует число его вхождений в тексте (можно вручную, можно с помощью `collections.Counter`)
  - НЕ нужно: проводить лемматизацию
- Выведите на экран **топ-20 самых частотных слов** (каждое слово напечатайте на новой строке)

```python
### РЕШЕНИЕ ###

import re
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

with open("ВойнаИМир10.txt", "r", encoding="utf-8") as f:
    text = f.read()

tokens = []
for token in word_tokenize(text.lower()):
    if (token not in stopwords.words("russian")) and (re.search("[А-ЯЁа-яёA-Za-z]", token)):
        tokens.append(token)
freq_dict = Counter(tokens)

for word, n in freq_dict.most_common(20):
    print(word)
```

## Задача 13

- Откройте в Python текст первых 10 глав романа «Война и мир» (ссылка)
- Смените регистр всех символов на нижний
- Удалите из списка токенов стоп-слова (используйте список стоп-слов из `nltk`)
- Удалите из списка токенов знаки препинания (подсказка: это можно сделать с помощью регулярных выражений)
- Проведите лемматизацию всего текста с помощью `pymorphy3`
- Выведите на экран через пробел все словоформы глагола ***идти***, которые встречаются в тексте (например: `"идущая шли идёт…"`)

*В ячейку с решением вставьте весь код, который вы написали при выполнении задания.*

*Если для запуска решения необходимы модули, включите импорт модулей в решение.*

```python
### РЕШЕНИЕ ###

import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pymorphy3

my_analyzer = pymorphy3.MorphAnalyzer()

with open("ВойнаИМир10.txt", "r", encoding="utf-8") as f:
    text = f.read()

idti_forms = []
for token in word_tokenize(text.lower()):
    if (token not in stopwords.words("russian")) and (re.search("[А-ЯЁа-яёA-Za-z]", token)):
        token_parse = my_analyzer.parse(token)[0]
        if token_parse.normal_form == "идти":
            idti_forms.append(token)

print(*idti_forms, sep=" ")
```

## Задача 14

- Откройте в Python текст первых 10 глав романа «Война и мир» (ссылка)
- Смените регистр всех символов на нижний
- Удалите из списка токенов стоп-слова (используйте список стоп-слов из `nltk`)
- Удалите из списка токенов знаки препинания (подсказка: это можно сделать с помощью регулярных выражений)
- Проведите лемматизацию всего текста с помощью `pymorphy3`
- Подсчитайте и выведите на экран **число всех существительных женского рода**, встречающихся в тексте
- Подсчитайте и выведите на экран **число всех существительных мужского рода**, встречающихся в тексте
  - (Подсказка: тег `pymorphy3` для существительных: `"NOUN"`, для женского рода: `"femn"`, для мужского рода: `"masc"`)

*В ячейку с решением вставьте весь код, который вы написали при выполнении задания.*

*Если для запуска решения необходимы модули, включите импорт модулей в решение.*

```python
### РЕШЕНИЕ ###

import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pymorphy3

my_analyzer = pymorphy3.MorphAnalyzer()

with open("ВойнаИМир10.txt", "r", encoding="utf-8") as f:
    text = f.read()

results = {"f": [], "m": []}
for token in word_tokenize(text.lower()):
    if (token not in stopwords.words("russian")) and (re.search("[А-ЯЁа-яёA-Za-z]", token)):
        token_parse = my_analyzer.parse(token)[0]
        if "NOUN" in token_parse.tag:
            if "femn" in token_parse.tag:
                results["f"].append(token)
            elif "masc" in token_parse.tag:
                results["m"].append(token)

print("femn:", len(results["f"]))
print("masc:", len(results["m"]))
```
